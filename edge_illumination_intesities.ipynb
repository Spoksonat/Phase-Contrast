```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#raw_0 = np.rot90(np.genfromtxt("1. RAW/RAW_hoja0.txt"),1)
raw_0 = np.rot90(np.genfromtxt("1. RAW/RAW_cilindro0.txt"),1)
mean_raw = np.mean(raw_0)
for i in range(256):
    for j in range(256):
        if raw_0[i,j] == 0:
            raw_0[i,j] = mean_raw

plt.imshow(raw_0, cmap="bone")

```

```python
ff_0 = np.rot90(np.genfromtxt("2. FF/FF_cilindro0.txt"),1)
mean_ff = np.mean(ff_0)
for i in range(256):
    for j in range(256):
        if ff_0[i,j] == 0:
            ff_0[i,j] = mean_ff

plt.imshow(ff_0, cmap="bone")
```

```python
img_0 = ff_0/raw_0
plt.imshow(img_0, cmap="bone")
```

```python
even = img_0[:,::2]
odd = img_0[:,1::2]

phase = (even-odd)/(even+odd)

plt.figure(figsize=(15,10))
plt.imshow(phase, cmap="bone")
plt.colorbar()
```

```python
perfiles = phase[0,:]
for i in range(1,256): 
    perfiles += phase[i,:]

plt.plot(perfiles)
```
